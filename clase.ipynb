```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Definir el intervalo
x = np.linspace(0, 1, 1000)

# Calcular las funciones de densidad
pdf_beta_1_1 = beta.pdf(x, 1, 1)
pdf_beta_12_17 = beta.pdf(x, 12, 17)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_beta_1_1, label='Prior', color='red', linestyle='--')
plt.plot(x, pdf_beta_12_17, label='Likelihood', color='black', linestyle='--', linewidth=3)
plt.plot(x, pdf_beta_12_17, label='Posterior', color='green', linewidth=4, alpha=0.3)


# Añadir títulos y etiquetas
plt.title('Exercise')
plt.xlabel('p')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```
